## This one will be main:
```python
#Removing odd numbers
for num in Fibo[:]:
  if num%2 != 0:
    Fibo.remove(num)  
```
What this is saying is take the list Fibo, and start at the first element till the last element and do as I say below. Well you might be tempted to say that `Fibo[:]:` is the same as `Fibo:`. Which is right but also wrong. Let me explain
```python
for num in Fibo:
```
The above snippet says that, "See python, there's a list called 'Fibo', and I want you to go through it's elements one by one and see if each num is not divisible by 2"... However in the later:
```python
for num in Fibo[:]:
```
you are saying to python, "Hey go through each element in the list, without any order, that is even if you have earlier read that element and it wasn't divisible by 2, just remove it...repeat this until you get all are divisible by 2(even)."
